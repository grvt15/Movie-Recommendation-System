```python
import ast

def extract_genres(text):
    genres = ast.literal_eval(text)
    return " ".join([genre['name'] for genre in genres])

dataframe['genres'] = dataframe['genres'].apply(extract_genres)
```

```python
dataframe['tags'] = (
    (dataframe['genres'] + ' ') * 4 +
    (dataframe['title'] + ' ') * 2 +
    dataframe['overview'] + ' ' +
    dataframe['tagline']
)
```

```python
import nltk
import re
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
import re

def preprocess(text):
    # Lowercase
    text = text.lower()

    # Remove punctuation
    text = re.sub(r'[^\w\s]', ' ', text)

    # Remove numbers
    text = re.sub(r'\d+', ' ', text)

    # Tokenize
    words = text.split()

    # Remove stopwords and lemmatize
    words = [
        lemmatizer.lemmatize(word)
        for word in words
        if word not in stop_words
    ]

    return " ".join(words)


dataframe['tags'] = dataframe['tags'].apply(preprocess)
```

```python
dataframe.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,stop_words='english', ngram_range=(1,2))
```

```python
dataframe = dataframe.reset_index(drop=True)

indices = pd.Series(
    data=dataframe.index,
    index=dataframe["title"]
)

# Keep only the first occurrence of duplicate titles
indices = indices[~indices.index.duplicated(keep="first")]

tfidf_matrix = tfidf.fit_transform(dataframe['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

%pip install rapidfuzz
```

```python
from rapidfuzz import process
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

def recommend(title, n=10):
    # Find the closest matching title
    match = process.extractOne(title, indices.index)

    if match is None or match[1] < 60:
        return ["Movie not found"]

    movie_title = match[0]

    idx = indices[movie_title]

    # Handle duplicate titles
    if isinstance(idx, pd.Series):
        idx = idx.iloc[0]

    sim_scores = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()

    similar_idx = sim_scores.argsort()[::-1][1:n+1]

    return dataframe.iloc[similar_idx]['title'].tolist()
```

```python
recommend("Toy Story")
```

```python
import pickle

# Save TF-IDF matrix
pickle.dump(tfidf_matrix, open('tfidf_matrix.pkl', 'wb'))

# Save title-to-index mapping
pickle.dump(indices, open('indices.pkl', 'wb'))

# Save DataFrame
dataframe.to_pickle('dataframe.pkl')

# Save TF-IDF vectorizer
pickle.dump(tfidf, open('tfidf.pkl', 'wb'))
```
